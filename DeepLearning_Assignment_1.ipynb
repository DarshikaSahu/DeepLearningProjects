```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
auto_mpg = fetch_ucirepo(id=9)

# data (as pandas dataframes)
X = auto_mpg.data.features
y = auto_mpg.data.targets

# metadata
print(auto_mpg.metadata)

# variable information
print(auto_mpg.variables)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(keras.layers.Dense(units=500, activation='relu', input_shape=(7,)))
model.add(keras.layers.Dense(units=250, activation='relu'))
model.add(keras.layers.Dense(units=125, activation='relu'))
model.add(keras.layers.Dense(units=1))
```

```python
model.summary()
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=False
)
```

```python
model.compile(
    optimizer='adam',
    loss='MeanSquaredError',
    metrics=['R2Score']
)
```

```python
model.fit(x=X_train, y=y_train, epochs=150, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)
mae = mean_absolute_error(y_test, pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
print("MAE:", mae)
```
